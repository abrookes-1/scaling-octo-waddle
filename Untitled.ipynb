```python
import requests
import json
```

```python
base_url = 'https://www.albion-online-data.com/api/v1/stats/prices/'
city = 'Lymhurst'

# T4_METALBAR_LEVEL1@1
# T5_METALBAR
# T8_PLANKS_LEVEL1@1
# T8_PLANKS_LEVEL3@3

items = {
    'bar': 'METALBAR',
    'plank': 'PLANKS',
}

levels = {
    3: 'T3_',
    4: 'T4_',
    5: 'T5_',
    6: 'T6_',
    7: 'T7_',
    8: 'T8_',
}

enchantments = {
    1: '_LEVEL1@1',
    2: '_LEVEL2@2',
    3: '_LEVEL3@3',
}

with open('items.txt', 'r') as itemsfile:
    item_list = json.load(itemsfile)

# print(item_list)
```

```python
PARAMS = {}

# URL = base_url + levels[4] + items['bar'] + enchantments[3]
# print(URL)
city = 'Caerleon'
threshold = 1000

i = -1
for key in item_list.keys():
    i += 1
    if i>1000:
        break
        
    city_price = 0
    bm_price = 0

    URL = base_url + key
    r = requests.get(url = URL, params = PARAMS)
    data = r.json()
    for city_data in data:
        if city_data['city'] == city:
            city_price = city_data['sell_price_min']
        elif city_data['city'] == 'Black Market':
            bm_price = city_data['buy_price_max']

    print(bm_price)
    if city_price != 0 and bm_price != 0:
        profit = bm_price - city_price
        print(profit)
        if profit > threshold:
            # profit!
            print(f'{item_list[key]} | {profit}')
```
